# **1. Data Ingestion**


```python
import pandas as pd
from llama_index.core import Document

# Đọc file CSV
df = pd.read_csv('/content/train.csv')

# Chuyển đổi từng hàng của CSV thành Document của LlamaIndex
documents = []

for _, row in df.iterrows():
    document_content = f"""
    Context: {row['context']}
    Article: {row['article']}
    Question: {row['question']}
    Extractive Answer: {row['extractive answer']}
    Abstractive Answer: {row['abstractive answer']}
    """
    documents.append(Document(text=document_content))
```

```python
from llama_index.core.node_parser import SentenceSplitter
from llama_index.core import Settings, VectorStoreIndex

text_splitter = SentenceSplitter(chunk_size=1028, chunk_overlap=128)
processed_documents = text_splitter(documents)

print("Before splitting: ", len(documents))
print("After splitting: ", len(processed_documents))
```

```python
print(processed_documents[1].text)
```

```python
from llama_index.core.node_parser import (
    SentenceSplitter,
    SemanticSplitterNodeParser,
)
from llama_index.embeddings.openai import OpenAIEmbedding
from llama_index.core import SimpleDirectoryReader
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

import os
import torch

device_type = torch.device("cuda" if torch.cuda.is_available() else "cpu")
embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-small-en-v1.5", device=device_type) # must be the same as the previous stage
splitter = SemanticSplitterNodeParser(
    buffer_size=1, breakpoint_percentile_threshold=95, embed_model=embed_model
)


nodes = splitter.get_nodes_from_documents(documents)
```

```python
from llama_index.core import (
    PromptTemplate,
    ChatPromptTemplate,
    SimpleDirectoryReader,

)
from llama_index.core.llms import ChatMessage, MessageRole


# 1️⃣ **System Prompt**
system_prompt = """
You are a multilingual and highly intelligent Q&A system that is trusted around the world.
Your task is to answer user queries based on the provided context and the following rules:
1. **Language rules**: If the query is in Vietnamese, answer in Vietnamese. If the query is in English, answer in English.
2. **Contextual answering rules**: Always answer using the provided context and not prior knowledge.
3. **Handling incomplete questions**: If the user's query is ambiguous or lacks key details, ask a clear and specific follow-up question to get the missing information.
4. **Answering style**: Be clear, concise, and avoid technical jargon unless necessary.
5. **Avoid explicit reference to the context**: Do not use phrases like 'Based on the context...' in your response.
"""

# 2️⃣ **User Prompt**
user_prompt = """
Context information is below.
---------------------
{context_str}
---------------------
Given the context information and not prior knowledge, answer the query.
Query: {query_str}
Answer:
"""

# 3️⃣ **Refine Prompt**
refine_prompt = """
The original query is as follows: {query_str}
We have provided an existing answer: {existing_answer}
We have the opportunity to refine the existing answer (only if needed) with some more context below.
------------
{context_msg}
------------
Given the new context, refine the original answer to better answer the query. If the context isn't useful, return the original answer.
Refined Answer:
"""

# 4️⃣ **Create Chat Prompt Template for System Instructions**
chat_prompt_template = ChatPromptTemplate(message_templates=[
    ChatMessage(content=system_prompt, role=MessageRole.SYSTEM),
    ChatMessage(content=user_prompt, role=MessageRole.USER)
])

# 5️⃣ **Create Prompt Templates for User and Refine Prompts**
user_prompt_template = PromptTemplate(user_prompt)
refine_prompt_template = PromptTemplate(refine_prompt)

# 6️⃣ **Initialize LLM (e.g., OpenAI, LangChain LLM, etc.)**

# 7️⃣ **Settings for LlamaIndex**

# 8️⃣ **Create the index with the new system (use Settings instead of ServiceContext)**
# Load documents from the ./data folder (replace with your data directory)

# Create the index with the prompts and LLM
index = VectorStoreIndex(
    documents,
    prompt_template=user_prompt_template,
    refine_prompt_template=refine_prompt_template,
    chat_prompt_template=chat_prompt_template,  # New chat prompt for system instructions
    embed_model=embed_model
)
```
